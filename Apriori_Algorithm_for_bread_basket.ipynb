```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/bread basket.csv')
df
```

```python
grp = df.groupby('Transaction')['Item'].apply(list).reset_index()
expand = grp['Item'].apply(pd.Series)
expand
```

```python
transaction = []
for i in range(0,len(expand)):
  transaction.append([str(expand.values[i,j]) for j in range(0,11)if str(expand.values[i,j]!='0')])
transaction
```

```python
!pip install apyori
```

```python
from apyori import apriori
apr = apriori(transactions=transaction,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
results = list(apr)
results
```

```python
def inspect(results):
  lhs = [tuple(i[2][0][0])[0] for i in results]
  rhs = [tuple(i[2][0][1])[0] for i in results]
  conf = [i[2][0][2] for i in results]
  lift = [i[2][0][3] for i in results]
  supp = [i[1] for i in results]
  return list(zip(lhs,rhs,conf,lift,supp))
df1 = pd.DataFrame(inspect(results),columns = ["item1","item2 ","conf","lift","supp"])
```

```python
df1.nlargest(10,columns='lift')
```
